## 7.문자열 포매팅

## 10.format()함수를 사용한 포매팅

```python
# 숫자 바로 대입
'I eat {} apples.'.format(3)
```

```python
# 문자 바로 대입
'I eat {} apples.'.format('three')
```

```python
# 변수로 대입
num = 3
'I eat {} apples.'.format(num)
```

```python
'I eat {0} apples.'.format('three')
```

```python
# 2개 이상 값 넣기
num = 10
day = 'three'
```

```python
'I ate {} apples. so I was sick for {} days'.format(num, day)
```

```python
'I ate {0} apples.so I was sick for {1}.'.format(num, day)
```

```python
'I ate {1} apples.so I was sick for {0} days.'.format(num,day)
```

```python
# 이름으로 넣기
'I ate {number} apples.so I was sick for {day} days.'.format(day=3, number=10)
```

```python
'I ate {} apples.so I was sick for {day} days.'.format(10, day=3)
```

```python
# 정렬
'{}'.format('hi')
```

```python
'{:<10}'.format('hi')
```

```python
'{:<10}'.format('hi')
```

```python
'{:<10}'.format('hi')
```

```python
#공백 채우기
'{:=^10}'.format('hi')
```

```python
'{:!<10}'.format('hi')
```

```python
#소수점 표현
'{:0.2f}'.format(3.428524)
```

```python
'{:10.2f}'.format(3.428524)
```

```python
'{{and}}'.format()
```

```python
'{}'.format('(and)')
```

## 11.f문자열 포매팅

```python
# 문자열 앞에 f접두사를 붙이기
name = '서창환'
age = 25
f'나의 이름은 {name}입니다.나이는 {age}입니다.'
```

```python
# 표현식 사용하기
f'나는 내년이면 {age +1}살이 됩니다.'
```

```python
d = {'name':'서창환', 'age': 26}
f'나의 이름은 {d['name']}입니다.나이는 {d['age']}입니다.'
```

```python
# 정렬
f'{'hi':<10}'
```

```python
f'{'hi':>10}'
```

```python
f'{'hi':^10}'
```

```python
f'{'hi':=^10}'
```

```python
f'{'hi':!<10}'
```

```python
# 소수점 표현
f'{3.4284231:0.2f}'
```

```python
f'{3.4284231:10.2f}'
```

```python
f'{3.4284231:0<10.2f}'
```

```python
f'{{and}}'
```
